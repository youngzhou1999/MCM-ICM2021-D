```python
df_edge = pd.read_csv('edge-rank.csv')
df_edge = df_edge.drop('Unnamed: 0', axis = 1)
df_ve_edge = df_edge.to_numpy()
df_ve_edge
```

```python
genre = [
 'Jazz',
 'Vocal',
 'Blues',
 'Religious',
 'Pop/Rock',
 'Country',
 'Folk',
 'R&B;',
 'Classical',
 'Electronic',
 'Stage & Screen',
 'Comedy/Spoken',
 'Avant-Garde',
 'International',
 'Reggae',
 'Latin',
 'Easy Listening',
 'New Age',
 "Children's",
 'Unknown'
]
```

```python
time = [1930,1940,1950,1960,1970,1980,1990,2000,2010]
my_genre = 'Country'
time_genre_mygenre = {}
for i in range(len(time)):
    time_i = str(time[i])
    time_genre_mygenre[str(time_i)] = {}
    for j in range(len(genre)):
        genre_j = genre[j]
        time_genre_mygenre[time_i][genre_j] = {}
        for k in range(len(genre)):
            genre_k = genre[k]
            time_genre_mygenre[time_i][genre_j][genre_k] = 0
```

```python
for i in range(len(df_ve_edge)):
    genre_i = df_ve_edge[i,2]
    genre_f = df_ve_edge[i,6]
    w = df_ve_edge[i,10]
    time_ed = str(df_ve_edge[i,7])
    time_genre_mygenre[time_ed][genre_f][genre_i] += w
```

```python
my_genre = 'Country'
time_inf = []
for i in range(len(time)):
    time_i = str(time[i])
    tmp = []
    for j in range(len(genre)):
        tmp.append((genre[j],time_genre_mygenre[time_i][my_genre][genre[j]]))
    time_inf.append(tmp)
```

```python

for i in range(len(time_inf)):
    time_inf[i].sort(key = lambda x:x[1], reverse = True)
```
